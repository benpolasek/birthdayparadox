```python
import random
duplicate_days = []
```

Generate a list of length equal to the number of students, where each entry represents the birthday of one student

```python
def random_bdays(n):
    """Returns a list of integers between 1 and 365, with length (n)."""
    t = []
    for i in range(n):
        bday = random.randint(1, 365)
        t.append(bday)
    return t
```

We now go through each 'class' of n students and see how many birthdays appear more than once and return that number as well as print the days that are shared, if any. We do this m times

```python
def count_matches(students, samples):
    """Generates (samples) samples of (students) students, and counts
    how many of them have at least one pair of students with the same bday."""
    count = 0
    for i in range(samples):
        t = random_bdays(students) 
        if has_duplicates(t):
            count += 1
            print('Duplicates days for interation ', i, ' are: ', duplicate_days)#printing out results of simulation
            del duplicate_days[:]
        else:
            print('There are no duplicate birthdays for iteration ', i)
    return (count)
```

If two or more students share the same birthday we add it to a list and return that list

```python
def has_duplicates(t):
    """Returns True if any element appears more than once in (t),
    False otherwise."""
    s = t[:]
    s.sort()
    k = 0
    for i in range(len(s)-1): #If any two birthdates are repeated we add that date to the list and return True, 
                                #that there was a repetion once all repeated dates have been added
        if s[i] == s[i+1]:
            duplicate_days.append(s[i])
    if len(duplicate_days) != 0:
        return True
    return False
```

Putting this all together, we print all the iterations and which birthdays were repeated. Then we calculate the probability that any two students will share a birthday based on the simulation of the number of students we input

```python
"""run the birthday simulation 1000 times and print the number of matches"""
num_students = 45
num_simulations = 1000
count = count_matches(num_students, num_simulations)
prob = count/num_simulations
print ('After %d simulations' % num_simulations)
print ('with %d students' % num_students)
print ('there were %d simulations with at least one match' % count)
print ('The probability that any two students share a birthday is: ', prob)
```
